```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('MICRODADOS.csv', sep=';',
                 encoding='latin-1', low_memory=False)
print(f'Total de registros: {len(df):,}')
print(f'Colunas: {list(df.columns)}')
df.head()
```

```python
# Exercício 1 - Visão Geral do Dataset
'''Exiba: (a) o número total de registros e colunas, 
(b) os tipos de dados de cada coluna com dtypes, 
(c) a quantidade e percentual de valores nulos por coluna (mostrando apenas as que têm nulos). 
Use shape, dtypes e isnull().sum().'''

# a) Número total de registros e colunas
num_registros, num_colunas = df.shape
print(f'Total de registros: {num_registros:,}')
print(f'Total de colunas: {num_colunas}')

# b) Tipos de dados de cada coluna
print('\nTipos de dados por coluna:')
print(df.dtypes)

# c) Quantidade e percentual de valores nulos por coluna
nulos = df.isnull().sum()
percentual_nulos = (nulos / len(df)) * 100
nulos_df = pd.DataFrame({'Quantidade de Nulos': nulos,
                        'Percentual de Nulos (%)': percentual_nulos})
nulos_df = nulos_df[nulos_df['Quantidade de Nulos'] > 0]
print('\nColunas com valores nulos:')
```

```python
# Exercício 2 - Distribuição por Classificação
# Utilizando a coluna Classificacao,
# calcule a frequência absoluta e percentual de cada categoria
# (Confirmados, Descartados, Suspeito, Síndrome Gripal não Especificada).
# Gere um gráfico de barras horizontal com os resultados.

frequencia = df['Classificacao'].value_counts()
percentual = df['Classificacao'].value_counts(normalize=True) * 100
classificacao_df = pd.DataFrame({'Frequência Absoluta': frequencia,
                                'Frequência Percentual (%)': percentual})
print(classificacao_df)

# Gráfico de barras horizontal
plt.figure(figsize=(10, 6))
classificacao_df['Frequência Absoluta'].plot(kind='barh', color='skyblue')
plt.title('Distribuição por Classificação')
plt.xlabel('Frequência Absoluta')
plt.ylabel('Classificação')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Exercício 3 - Top 10 Municípios com Mais Notificações
# Identifique os 10 municípios com mais notificações no dataset.
# Gere um gráfico de barras mostrando o ranking.
# Escreva em Markdown qual município lidera e quantos registros possui.

top_municipios = df['Municipio'].value_counts().head(10)
print(top_municipios)

# Gráfico de barras
plt.figure(figsize=(10, 6))
top_municipios.plot(kind='bar', color='salmon')
plt.title('Top 10 Municípios com Mais Notificações')
plt.xlabel('Município')
plt.ylabel('Número de Notificações')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Exercício 4 - Distribuição por Sexo
# Calcule a distribuição de notificações por Sexo (F, M, I).
# Gere um gráfico de pizza (pie chart) com percentuais.
# Interprete em Markdown: qual sexo concentra mais notificações?

sexo_frequencia = df['Sexo'].value_counts()
print(sexo_frequencia)

# Gráfico de pizza
plt.figure(figsize=(8, 8))
sexo_frequencia.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=[
                     'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Distribuição por Sexo')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

```python
# Exercício 5 - Casos por Faixa Etária
#  Usando a coluna FaixaEtaria, gere um gráfico de barras mostrando o número de notificações por faixa etária.
# Ordene da menor para a maior faixa.
# Qual faixa concentra o maior volume?

faixa_frequencia = df['FaixaEtaria'].value_counts().sort_index()
print(faixa_frequencia)

# Gráfico de barras
plt.figure(figsize=(10, 6))
faixa_frequencia.plot(kind='bar', color='mediumpurple')
plt.title('Número de Notificações por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Notificações')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Exercício 6 - Taxa de Letalidade
# Filtre apenas os casos Confirmados (Classificacao == 'Confirmados').
# Calcule quantos evoluíram para Óbito pelo COVID-19, Óbito por outras causas, Cura e Ignorado.
# Calcule a taxa de letalidade: (óbitos COVID / total de confirmados) × 100.
# Exiba os resultados numéricos e a taxa.

confirmados = df[df['Classificacao'] == 'Confirmados']
obitos_covid = confirmados[confirmados['Evolucao']
                           == 'Óbito pelo COVID-19'].shape[0]
obitos_outras = confirmados[confirmados['Evolucao']
                            == 'Óbito por outras causas'].shape[0]
cura = confirmados[confirmados['Evolucao'] == 'Cura'].shape[0]
ignorado = confirmados[confirmados['Evolucao'] == 'Ignorado'].shape[0]
total_confirmados = confirmados.shape[0]
taxa_letalidade = (obitos_covid / total_confirmados) * 100
print(f'Óbitos pelo COVID-19: {obitos_covid}')
print(f'Óbitos por outras causas: {obitos_outras}')
print(f'Cura: {cura}')
print(f'Ignorado: {ignorado}')
print(f'Taxa de Letalidade: {taxa_letalidade:.2f}%')
```

```python
# Exercício 7 - Sintomas mais Frequentes
# As colunas de sintomas são: Febre, DificuldadeRespiratoria, Tosse, Coriza, DorGarganta, Diarreia e Cefaleia.
# Para cada uma, conte quantos registros têm valor "Sim".
# Gere um gráfico de barras horizontal ordenado do mais ao menos frequente.

sintomas = ['Febre', 'DificuldadeRespiratoria', 'Tosse',
            'Coriza', 'DorGarganta', 'Diarreia', 'Cefaleia']

sintomas_frequencia = {
    sintoma: df[df[sintoma] == 'Sim'].shape[0] for sintoma in sintomas}

sintomas_frequencia = dict(
    sorted(sintomas_frequencia.items(), key=lambda item: item[1], reverse=True))
print(sintomas_frequencia)

# Gráfico de barras horizontal
plt.figure(figsize=(10, 6))
plt.barh(list(sintomas_frequencia.keys()),
         list(sintomas_frequencia.values()), color='lightseagreen')
plt.title('Sintomas mais Frequentes')
plt.xlabel('Número de Registros com "Sim"')
plt.ylabel('Sintoma')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Exercício 8 - Comorbidades nos Óbitos por COVID
# Filtre os registros com Evolucao == 'Óbito pelo COVID-19'.
# Para as colunas ComorbidadePulmao, ComorbidadeCardio, ComorbidadeRenal, ComorbidadeDiabetes, ComorbidadeTabagismo e ComorbidadeObesidade,
# conte quantos óbitos tinham cada comorbidade ("Sim").
# Gere um gráfico de barras e interprete em Markdown: qual comorbidade está mais presente nos óbitos?

comorbidades = ['ComorbidadePulmao', 'ComorbidadeCardio', 'ComorbidadeRenal',
                'ComorbidadeDiabetes', 'ComorbidadeTabagismo', 'ComorbidadeObesidade']
obitos_covid_df = df[df['Evolucao'] == 'Óbito pelo COVID-19']
comorbidades_frequencia = {
    comorbidade: obitos_covid_df[obitos_covid_df[comorbidade] == 'Sim'].shape[0] for comorbidade in comorbidades}
comorbidades_frequencia = dict(
    sorted(comorbidades_frequencia.items(), key=lambda item: item[1], reverse=True))
print(comorbidades_frequencia)

# Gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(list(comorbidades_frequencia.keys()), list(
    comorbidades_frequencia.values()), color='orchid')
plt.title('Comorbidades nos Óbitos por COVID-19')
plt.xlabel('Comorbidade')
plt.ylabel('Número de Óbitos com "Sim"')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Exercício 9 - Evolução Temporal das Notificações
# Converta DataNotificacao para datetime com pd.to_datetime(df['DataNotificacao']).
# Crie uma coluna AnoMes usando .dt.to_period('M').
# Agrupe por AnoMes, conte as notificações e gere um gráfico de linha.
# Identifique em Markdown os períodos de pico (ondas da pandemia).

df['DataNotificacao'] = pd.to_datetime(df['DataNotificacao'], errors='coerce')
df['AnoMes'] = df['DataNotificacao'].dt.to_period('M')
notificacoes_por_ano_mes = df.groupby('AnoMes').size()
print(notificacoes_por_ano_mes)

# Gráfico de linha
plt.figure(figsize=(12, 6))
notificacoes_por_ano_mes.plot(kind='line', marker='o', color='steelblue')
plt.title('Evolução Temporal das Notificações')
plt.xlabel('Ano-Mês')
plt.ylabel('Número de Notificações')
plt.xticks(rotation=45)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
#  Exercício 10 - Tabela Cruzada (Pivot Table)
# Crie uma tabela cruzada com pd.crosstab entre os 5 municípios com mais casos confirmados e a coluna Evolucao.
# Exiba a tabela.
# Em seguida, calcule a taxa de letalidade por município (óbitos COVID / total confirmados × 100) e identifique qual dos 5 tem a maior taxa.

top_5_municipios = df['Municipio'].value_counts().head(5).index
crosstab = pd.crosstab(df[df['Municipio'].isin(
    top_5_municipios)]['Municipio'], df['Evolucao'])
print(crosstab)
letalidade_municipios = {}
for municipio in top_5_municipios:
    confirmados_municipio = crosstab.loc[municipio].sum()
    obitos_covid_municipio = crosstab.loc[municipio].get(
        'Óbito pelo COVID-19', 0)
    taxa_letalidade = (obitos_covid_municipio / confirmados_municipio) * \
        100 if confirmados_municipio > 0 else 0
    letalidade_municipios[municipio] = taxa_letalidade
letalidade_municipios = dict(
    sorted(letalidade_municipios.items(), key=lambda item: item[1], reverse=True))
print('\nTaxa de Letalidade por Município:')

for municipio, taxa in letalidade_municipios.items():
    print(f'{municipio}: {taxa:.2f}%')
```
